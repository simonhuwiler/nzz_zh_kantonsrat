```python
import pandas as pd
from pathlib import Path
import json
```

```python
df_votum_raw = pd.concat([
    pd.read_csv(Path('../export/votum/votum_0.csv')),
    pd.read_csv(Path('../export/votum/votum_1.csv'))
])

# Remove non members (mostly former members who are now in the Regierungsrat)
df_votum_raw = df_votum_raw[df_votum_raw.ismember == True]

# Typecast
df_votum_raw['sitzung_date'] = pd.to_datetime(df_votum_raw['sitzung_date'])
```

```python
df_sample = df_votum_raw[df_votum_raw.sitzung_date.between('2021-09-01', '2021-09-30')]

b = df_sample[['sitzung_date', 'geschlecht']].sort_values('sitzung_date')
#df_sample.to_csv(Path('../../nzz_zh_kantonsrat_genderstudy_sonofication/data/1.csv'), index=False, encoding='utf-8')

#df_sample.T

records = []
lastdate = df_sample['sitzung_date'].min()
for i, row in df_sample.iterrows():
    if lastdate != row['sitzung_date']:
        records.append('x')
        lastdate = row['sitzung_date']

    records.append(row['geschlecht'])

str(records)
```
