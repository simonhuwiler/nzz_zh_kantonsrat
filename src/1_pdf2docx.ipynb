```python
from pathlib import Path
from pdfminer.high_level import extract_pages, extract_text_to_fp
from pdfminer.layout import LTTextContainer, LTChar
from pdfminer.layout import LAParams
```

```python
f = '../export/Files/2021-09-27-17fa216f98fd4d38afd5afb53743ec65-332.pdf'

def read_pdf(f):
    i = 0

    textbuf = ""
    reading = False

    paragraphs = []

    fakes = ['Detailberatung', 'Titel und Ingress', 'I. und II.', 'III.', 'Abstimmung', 'IV.',
    'Schlussabstimmung']

    pages = extract_pages(f)
    for page_layout in pages:
        i += 1
        #if i < 23: continue
        #if i > 24: break

        #print(i)
        for element in page_layout:
            if isinstance(element, LTTextContainer):

                for text_line in element:

                    if reading:
                        textbuf += text_line.get_text()

                    if reading:
                        if len(text_line.get_text().strip()) == 0:
                            # Stop reading
                            paragraphs.append(textbuf)
                            reading = False
                            textbuf = ""
                            
                    else:
                        # not Reading. Check first Character
                        for character in text_line:
                            if isinstance(character, LTChar):

                                if 'italic' in character.fontname.lower():
                                    # Is Italic

                                    # Check, if not fake italic
                                    is_fake = False
                                    for fake in fakes:
                                        if fake in text_line.get_text():
                                            is_fake = True
                                            break

                                    if not is_fake:
                                        # Start reading
                                        textbuf = text_line.get_text()
                                        reading = True
                            break

    # If still reading, add to dict
    if reading:
        paragraphs.append(textbuf)

#for p in paragraphs:
#    print(p)
#    print("----")
#print(paragraphs)
print("finito")
```
